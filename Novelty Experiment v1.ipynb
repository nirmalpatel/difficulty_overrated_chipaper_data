# Novelty experiment in Battleship Numberline

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import statsmodels_helper as helper
pd.set_option('max_columns', 50)
plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/novelty_clean_v1.csv.gz')
```

```python
df.head()
```

```python
df['difficulty'] = 1 - df.meanAccuracy
df['log_levelLength'] = np.log(df.levelLength)
```

## totalTrials ~ log(levelLength) + log(levelLength)^2

```python
mod = smf.ols(formula='totalTrials ~ log_levelLength + I(log_levelLength**2)', data=df)
res = mod.fit()
df['predicted_totalTrials'] = res.fittedvalues
print(res.summary2())
```

```python
test_df = pd.DataFrame({'log_levelLength': np.log(np.arange(100) + 1), })
test_df['predicted_totalTrials'] = res.predict(test_df)
prstd, iv_l, iv_u = wls_prediction_std(res, helper.transform_exog_to_model(res, test_df))
```

```python
from matplotlib.ticker import FormatStrFormatter

plt.figure(figsize=(8,6))
plt.plot(np.arange(100)+1, test_df.predicted_totalTrials, lw=3)
plt.xscale('log')
plt.tick_params(axis='x', which='minor')
ax = plt.gca()
ax.xaxis.set_minor_formatter(FormatStrFormatter("%d"))
plt.xlabel('Level Length')
plt.ylabel('Engagement (# Trials Played)')
plt.title('Engagement v/s Level Length')
plt.ylim((10,40))
plt.scatter(df.levelLength, df.totalTrials)
plt.xlim((0,105))
plt.show()
```

## difficulty ~ log(levelLength) + log(levelLength)^2

```python
mod = smf.ols(formula='difficulty ~ log_levelLength + I(log_levelLength**2)', data=df)
res = mod.fit()
df['predicted_meanAccuracy'] = res.fittedvalues
print(res.summary2())
```

No p-values are significant when both log(levelLength) and log(levelLength)^2 are present, mostly due to interaction effect. All p-values are significant when we separate them out. Consider models below.

### difficulty ~ log(levelLength)

```python
mod1 = smf.ols(formula='difficulty ~ log_levelLength', data=df)
res1 = mod1.fit()
print(res1.summary2())
```

### difficulty ~ log(levelLength)^2

```python
mod2 = smf.ols(formula='difficulty ~ I(log_levelLength**2)', data=df)
res2 = mod2.fit()
print(res2.summary2())
```
