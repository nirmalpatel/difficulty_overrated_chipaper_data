```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
import seaborn as sns
import statsmodels_helper as helper
pd.set_option('max_columns', 50)
import itertools
```

```python
df = pd.read_csv('data/diff_choice_clean_v2.csv.gz')
```

```python
df.head()
```

## totalTrials ~ predicted_difficulty + predicted_difficulty_square + choice + interactions

```python
mod = smf.ols(formula='totalTrials ~ predicted_mean_difficulty + I(predicted_mean_difficulty**2) + choice + \
              predicted_mean_difficulty*choice + I(predicted_mean_difficulty**2)*choice', data=df)
res = mod.fit()
df['predicted_totalTrials'] = res.fittedvalues
print(res.summary())
```

```python
# Here we make a test data frame for which we will predict values

test_dict = {'predicted_mean_difficulty': np.linspace(.3,.9,num=100), 'choice':['choice','no choice'] }
test_df = pd.DataFrame(list(itertools.product(*test_dict.values())), columns=test_dict.keys())
test_df.head()
```

```python
sns.lmplot("predicted_mean_difficulty", "predicted_totalTrials", hue="choice", data=df, order=2, ci=95)
plt.ylabel('predicted total trials')
```

```python
sns.countplot("difficulty", hue="pretestScore", data=df.query('condition == "choice-feedforward" & \
            (domain == "fraction" | domain == "whole")'), 
             order=['1_Very_Easy', '2_Easy', '3_Medium', '4_Hard', '5_Very_Hard'], hue_order = ['low', 'high'])
```

## engagement ~ condition + difficulty

```python
mod = smf.ols(formula='log_totalTrials_totalTime ~ condition + difficulty', data=df)
res = mod.fit()
df['predicted_engagement'] = res.fittedvalues
print(res.summary())
```

```python
df['persistence'] = df.log_totalTrials_totalTime - df.predicted_engagement
```

```python
sns.distplot(df.persistence, kde=False)
```

```python
sns.factorplot(x='difficulty', y='persistence', col='feedforward', row='choice', data=df, 
               order=['1_Very_Easy', '2_Easy', '3_Medium', '4_Hard', '5_Very_Hard'])
```
